```python
# import packages
import numpy as np
import plotly.graph_objects as go
import pandas as pd
```

```python
class LoadAero:
    """ the class gets a .dat file and plots the distribution, plus other info """

    def __init__(self, filename):
        self.file = filename

    def get_mat(self):

        mat = np.zeros((81, 41))

        with open(self.file) as f:
            row = 0
            for line in f:
                values = line.split(",")
                values = [float(i) for i in values]
                mat[row, :] = values
                row += 1

        return mat

    def get_shape(self):
        return self.get_mat().shape

    def plot_distribution(self):

        return 0
```

```python
a320 = LoadAero("load.dat")
mat = a320.get_mat()
shape = a320.get_shape()

shape
```

```python
mat
```

```python
C_a = 0.547
l_a = 2.771

N_z, N_x = shape[0], shape[1] 

theta_z = np.arange(N_z + 1)*(np.pi/N_z)
theta_x = np.arange(N_x + 1)*(np.pi/N_x)

z_coord = np.zeros(N_z)
x_coord = np.zeros(N_x)

def get_coord(theta, theta_1, seg):
    coord = -0.5*((seg/2)*(1-np.cos(theta)) + (seg/2)*(1-np.cos(theta_1)))
    return coord

for i in range(N_z):
    z_coord[i] = get_coord(theta_z[i], theta_z[i+1], C_a)
    
for i in range(N_x):
    x_coord[i] = get_coord(theta_x[i], theta_x[i+1], l_a)

print(z_coord)
print(x_coord)
```

```python
df = pd.DataFrame(mat, z_coord)
df.columns = x_coord

df
```

```python
fig = go.Figure(data=[go.Surface(z=df.values, x=x_coord, y=z_coord)])

fig.show()
```
